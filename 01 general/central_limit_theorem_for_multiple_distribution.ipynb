### For exponential distribution

Even the population is a exponential distribution, its sample mean is normal distribution.

```python
import matplotlib.pyplot as plt
import random
import numpy as np

num_samples = 1000

random_numbers = np.random.exponential(1, num_samples)

plt.hist(random_numbers, bins=20, density=True)

x = np.linspace(0, 5, 100)
pdf = np.exp(-x)
plt.plot(x, pdf, '--', color='r', label='PDF')

plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

#plt.savefig('histogram_pdf.png')
plt.show()
```

```python
num_samples = 1000
sample_size = 50
num_iterations = 300

means = []

for i in range(num_iterations):
    sample = np.random.choice(random_numbers, sample_size)
    mean = np.mean(sample)
    means.append(mean)


x = np.linspace(np.mean(means)-3*np.std(means), np.mean(means)+3*np.std(means), 100)
pdf = np.exp(-(x-np.mean(means))**2/(2*(np.std(means))**2))/np.sqrt(2*np.pi*(np.std(means))**2)


plt.hist(means, bins=19, density=True)


plt.plot(x, pdf, '--', color='r', label='Normal PDF')

plt.title('Distribution of Sample Means')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.legend()

#plt.savefig('sample_means_histogram_pdf.png')
plt.show()
```
